```python
data = datasets.load_iris()
data.keys()
print(data["DESCR"])
```

```python
data["data"]
```

```python
df = pd.DataFrame(data["data"], columns=data["feature_names"])
```

```python
df["target"] = data["target"]
df.head()
```

```python
col = "sepal length (cm)"
sns.relplot(x=col, y="target", hue="target", data=df)
plt.suptitle(col,y=1.05)
```

```python
sns.pairplot(df, hue="target")
```
